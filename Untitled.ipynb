```python
import time
import datetime
import ssl
import numpy as np
import secret
import pymongo

client = pymongo.MongoClient(secret.connstring,
                             ssl=True,
                             ssl_cert_reqs=ssl.CERT_NONE) #connstring is a variable comming from secret.py containing mongo db connection string
db=client.test

import requests

#select unique dates

#select only those dates, where all digits are present

#loop over dates

#recreate integer value for gas reading

#send data to the server

humidity="None" #default value
temperature="None" #default value
url = 'http://35.234.91.159/gas/update.php'
```

```python
dates=db.gas_digit.distinct("date")
```

```python
def extract_intvalue(res):
    if(len(res)==8):
        reading=["","","","","","","",""]
        for p in res:
            reading[p["digit_position"]]=str(p["label"])
        joined="".join(reading)
        if(len(joined)!=8):
            raise ValueError()
    else:
        raise ValueError()
    return int(joined)
```

```python
#'round' timestamps in mongodatabase
for record in db.gas_digit.find():
    db.gas_digit.update_one(
        {"_id":record["_id"]},
        {"$set":{
            "date":datetime.datetime.fromtimestamp(
                round(record['date'].timestamp()/10)*10)
        }})
```

```python
for date in dates:
    if date>datetime.datetime(2020,6,13):
        res=list(db.gas_digit.find({"date":date},{"label":1,"date":1,"digit_position":1}))
        try:
            print(date,end ="  ")
            gas_reading=extract_intvalue(res)
            print(gas_reading,end="   ")
            url = 'http://35.234.91.159/gas/update.php'
            myobj = {'temperature': temperature,'gas':gas_reading,'humidity':humidity,'labelled':1,'time':date,'gas_processed':gas_reading}
            print(requests.post(url, data = myobj))
            time.sleep(0.2)
        except ValueError:
            print("value error")
        except:
            print("unexpected error")
```
